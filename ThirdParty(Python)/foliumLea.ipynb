#### folium.folium.Map()参数详解

```python
folium.Map(
    location=None,              # 地图中心, 纬度和经度
    width='100%',               # 地图宽度，如果是int则表示像素值，如果是str则表示百分比；
    height='100%',              # 地图高度
    left='0%',
    top='0%',
    position='relative',
    tiles='OpenStreetMap',       # 地图样式，默认为OpenStreetMap
    attr=None,                   # 当使用自定义瓦片时，传入自定义瓦片的名词
    min_zoom=0,                  # 地图可缩放的最小级别， 默认为0级
    max_zoom=18,                 # 地图可缩放的最小级别，默认为18级；
    zoom_start=10,               # 地图初始缩放级别，默认为10级；
    min_lat=-90,
    max_lat=90,
    min_lon=-180,
    max_lon=180,
    max_bounds=False,
    crs='EPSG3857',              # 投影坐标系标识
    control_scale=False,         # 是否在地图上添加比例尺，默认为False；
    prefer_canvas=False,         # 强制使用Canvas渲染
    no_touch=False,              # 是否禁止手动操作，默认为False；
    disable_3d=False,            # 强制使用CSS 3D效果
    png_enabled=False,
    zoom_control=True,           # 是否要限制zoom操作
)
```

##### 内建地图样式“tiles”参数可选值：
- "OpenStreetMap"
- "Mapbox Bright" (Limited levels of zoom for free tiles)
- "Mapbox Control Room" (Limited levels of zoom for free tiles)
- "Stamen" (Terrain, Toner, and Watercolor)
- "Cloudmade" (Must pass API key)
- "Mapbox" (Must pass API key)
- "CartoDB" (positron and dark_matter)

##### “crs”投影坐标系标识
- EPSG3857: Web墨卡托投影后的平面地图，坐标单位为米。大部分国外地图使用的时该标准。
- EPSG4326: Web墨卡托投影后的平面地图，但仍然使用WGS84的经度、纬度表示坐标。
- EPSG3395: 墨卡托投影，主要用于航海图

参数修改示例

```python
tsing_map = folium.Map([39.981771, 116.347313],
               zoom_start=15,
               control_scale=True
              )
tsing_map
```

高德卫星图

```python
tiles = 'https://webst02.is.autonavi.com/appmaptile?style=6&x={x}&y={y}&z={z}'
state_world = folium.Map([40.002694, 116.322373],
           tiles= tiles,
           attr='高德-卫星影像图',
           zoom_start=15,
          )
state_world
```

```python
import folium
 
bj_map = folium.Map(location=[39.93, 115.40], zoom_start=12)
 
folium.Marker(
    location=[39.95, 115.33],
    popup='Mt. Hood Meadows',
    icon=folium.Icon(icon='cloud')
).add_to(bj_map)
 
folium.Marker(
    location=[39.96, 115.32],
    popup='Timberline Lodge',
    icon=folium.Icon(color='green')
).add_to(bj_map)
 
folium.Marker(
    location=[39.93, 115.34],
    popup='Some Other Location',
    icon=folium.Icon(color='red', icon='info-sign')    # 标记颜色  图标
).add_to(bj_map)
 
bj_map
```

```python
bj_map = folium.Map(location=[39.93, 116.40], zoom_start=12)
 
folium.Circle(
    radius=200,
    location=(39.92, 116.43),
    popup='The Waterfront',
    color='#00FFFF',   # 颜色
    fill=False,        # 填充
).add_to(bj_map)
 
folium.CircleMarker(
    location=(39.93, 116.38),
    radius=50,   # 圆的半径
    popup='Laurelhurst Park',
    color='#FF1493',
    fill=True,
    fill_color='#FFD700'
).add_to(bj_map)
 

bj_map
```

```python
import folium
 
dynamic_tagging = folium.Map(
    location=[46.8527, -121.7649],
    zoom_start=13
)
 
folium.Marker(
    [46.8354, -121.7325],
    popup='Camp Muir'
).add_to(dynamic_tagging)
 
dynamic_tagging.add_child(folium.ClickForMarker(popup='Waypoint'))
dynamic_tagging
```
